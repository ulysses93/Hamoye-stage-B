### Removing dates and lights columns

```python
data1 = data.drop('lights',inplace=False, axis=1 )
```

```python
data2 = data1.drop('date',inplace=False, axis=1)
```

```python
data2
```

### Normalizing data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalised_data2 = pd.DataFrame(scaler.fit_transform(data2), columns=data2.columns)
#setting up features and targets
features_data2 = normalised_data2.drop(columns='Appliances')
Appliances_target= normalised_data2['Appliances']

#spliting our data into training and testing dataset

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(features_data2,Appliances_target,test_size=0.3,random_state=42)

from sklearn.linear_model import LinearRegression
model= LinearRegression()

#fit the model to training dataset

model.fit(x_train,y_train)
predicted_values = model.predict(x_test)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test,predicted_values)
round(mae, 3)
```

```python
#residual sum of squares
rss = np.sum(np.square(y_test-predicted_values))
round(rss, 3)
```

```python
#root mean square error
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse, 3)
```

```python
#r-squared

from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score,3)
```

### Regularizing 


```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train,y_train)
```

### Comparing the effects of regulirisation for feature selection

```python
def get_weights_data2(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_,feat.columns).sort_values()
    weights_data2 = pd.DataFrame(weights).reset_index()
    weights_data2.columns = ['Features', col_name]
    weights_data2[col_name].round(3)
    return weights_data2

linear_model_weights = get_weights_data2(model,x_train, 'Linear_Model_Weight')
ridge_weights_data2 = get_weights_data2(ridge_reg,x_train, 'Ridge_Weight')
lasso_weights_data2 = get_weights_data2(lasso_reg,x_train, 'Lasso_Weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_data2, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_data2, on='Features')
```
